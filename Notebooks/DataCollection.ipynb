### Code for collecting human written codes from SecurityEval Dataset

```python
import os
import shutil

file_counter = 0
insecure_code_path = r"SecurityEval-main/Testcases_Insecure_Code"
author_code_path = r"Author_Insecure_Code"

# List of folders in the insecure code path
cwe_folders = os.listdir(insecure_code_path)

for folder in cwe_folders:
    folder_path = os.path.join(insecure_code_path, folder)
    
    # Only process if it's a directory
    if os.path.isdir(folder_path):
        files = os.listdir(folder_path)
        
        # List of files that match the condition
        author_files = [file for file in files if file.endswith(".py") and file.startswith("author")]
        
        if author_files:
            file_counter += len(author_files)
            # Ensure destination directory exists
            destination_folder = os.path.join(author_code_path, folder)
            os.makedirs(destination_folder, exist_ok=True)
            
            # Copy only the "author" files
            for file in author_files:
                src = os.path.join(folder_path, file)
                dst = os.path.join(destination_folder, file)
                
                # If file doesn't exist, copy it
                if not os.path.exists(dst):
                    shutil.copy(src, dst)
                else:
                    print(f"File {dst} already exists. Skipping...")
```

```python
print(f"Total 'author' files copied: {file_counter}")
```
